Sales Analysis

```python
import pandas as pd
import os
# Importing pandas for data manipulation and os for file handling
```

Merging 12 months of data together into one file

```python
all_months_df = pd.DataFrame()

files = [file for file in os.listdir('./Sales_Data/')]


for file in files:
    df = pd.read_csv(f"""./Sales_Data/{file}""")
    
    all_months_df = pd.concat([all_months_df, df])
    
all_months_df.to_csv("all_data.csv", index=False)
# Reading data from a CSV file and displaying the first few rows
```

Read the merged df

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
# Reading data from a CSV file and displaying the first few rows
```

Cleanup the data

```python
all_data = all_data.dropna(how='all')
all_data.head()
# Removing rows with NaN values and displaying the cleaned data
```

```python
all_data.info()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

Augment data with additional month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
# Extract month from 'Order Date'
all_data['Month'] = all_data['Month'].astype('int')
# Convert 'Month' column to integer type
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
# Convert 'Quantity Ordered' to numeric
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
# Convert 'Price Each' to numeric
all_data.head()
# Display first few rows of the DataFrame
```

```python
# add total sales column
all_data['Sales'] = all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

```python
all_data.to_csv('all_data.csv', index=False)
```
